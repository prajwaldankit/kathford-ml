```python

def to_lowercase(name):
    return name.lower()

def greet(name):
    normalized_name = to_lowercase(name)
    return f"Hello {normalized_name}"

def normalized_func(func, value):
    return func(value)


print(normalized_func(greet, "ANKIT"))
```

```python
def logger(func):
    def wrapper(req):
        print("request recived")
        return func(req.upper())
    return wrapper

@logger
def fetch_homepage(string):
    return f"Homepage {string}"

fetch_homepage("hello")
```

```python
def nomalize(x):
    return x.strip()
def to_upper(x):
    return x.upper()

def pipeline(data, funcs):
    for func in funcs:
        data=func(data)
    return data

result = pipeline("   fdasfas    ", [nomalize, to_upper])
print("   fdasfas    ")
print(result)
```

```python
with open("example.csv", "r") as csv_file:
    result = []
    
    for line in csv_file:
        result.append(list(map(lambda x : x.strip(), line.split(','))))
    
    keys = result[0]
    values = result[1:]
    final_result = []

    for val in values:
        temp = {}
        for i in range(len(keys)):
            temp[keys[i]] = val[i]
        final_result.append(temp)

    print(final_result)
```
