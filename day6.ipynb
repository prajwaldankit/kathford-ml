**Students**
- name, roll, marks

```python
student_1_name = "ANkit"
student_1_roll = 1
student_1_marks = [60,40,50]

student_2_name = "Prajwal"
student_2_roll = 10
student_2_marks = [61,90,60]
```

```python
def calculate_avg_1(marks):
    return sum(marks) / len(marks)

student_1_avg = calculate_avg_1(student_1_marks)
student_2_avg = calculate_avg_1(student_2_marks)
print(student_1_name, student_1_avg)
print(student_2_name, student_2_avg)
```

```python
student_1 = {
    "name" : "Ankit",
    "roll": 1,
    "marks": [40,50,60] 
}

student_2 = {
    "name" : "PRajwal",
    "roll": 1,
    "marks": [40,80,60] 
}
```

```python
def calculate_avg_2(student):
    return sum(student["marks"]) / len(student["marks"])

student1_avg = calculate_avg_2(student_1)
student2_avg = calculate_avg_2(student_2)

print(student_1["name"], student1_avg)
print(student_2["name"], student2_avg)
```
