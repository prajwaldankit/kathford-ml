```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
points = pd.read_csv("data.csv")

plt.scatter(points.x, points.y)
plt.show()
```

```python
def loss_function(w,b,points):
    total_error = 0
    for i in range(len(points)):
        x = points.iloc[i].x
        y = points.iloc[i].y
        total_error += (y - (w*x + b)) ** 2
    return total_error / len(points)

print(loss_function(1.9,3, points))        
```

```python
def gradient_descent(w_now, b_now, points, learning_rate):
    w_gradient = 0
    b_gradient = 0

    n = len(points)

    for i in range(n):
        x = points.iloc[i].x
        y = points.iloc[i].y

        w_gradient += (-2/n) * x * (y - (w_now * x + b_now))
        b_gradient += (-2/n) * (y - (w_now * x + b_now))
    
    w = w_now - learning_rate * w_gradient
    b = b_now - learning_rate * b_gradient
    
    return w,b
```

```python
w = 0 
b = 0
learning_rate = 0.001
epochs = 1000

for i in range(epochs):
    w, b = gradient_descent(w, b, points, learning_rate)

print(w)
print(b)
```

```python
plt.scatter(points.x, points.y)
line_x = np.arange(00,60)
line_y = w * line_x + b
plt.plot(line_x, line_y, color="red")
plt.show()
```
