```python
class Student:
    def __init__(self, name, roll, marks):
        self.name = name
        self.roll = roll
        self.marks = marks
    
    def get_role(self):
        return "Student"

    def get_average(self):
        return sum(self.marks) / len(self.marks)
    
    def add_mark(self, mark):
        self.marks.append(mark)

    def is_pass(self):
        # Check each individual marks
        if self.get_average() >= 50:
            return True
        else:
            return False
        
    def display(self):
        return f"Name: {self.name} Roll: {self.roll}, Average: {self.get_average()}"
```

```python
student1 = Student("Ankit", 23, [20,40,25])
student2 = Student("Prajwal", 23, [65,74,93])
print(student1.name, student1.get_average())

print(student1.marks)
student1.add_mark(34)
print(student1.marks)
print(student1.get_average())
print(student1.is_pass())
print(student1.display())
```

```python
class GraduateStudent(Student):
    def __init__(self, name, roll, marks, thesis_topic):
        super().__init__(name, roll, marks)
        self.thesis_topic = thesis_topic

    def get_role(self):
        return "Graduate Student"
```
