```python
class Student:
    def __init__(self, name, roll, marks):
        self.name = name
        self.roll = roll
        self.marks = marks

    def get_role(self):
        return "Student"

    def get_average(self):
        return sum(self.marks) / len(self.marks)

    def add_mark(self, mark):
        self.marks.append(mark)

    def is_pass(self):
        if all(mark >= 40 for mark in self.marks):
            return False

        if self.get_average() >= 50:
            return True
        else:
            return False
        
    def predict_grade(self):
        average = self.get_average()
        if average >= 80:
            return "A"
        elif average >= 60:
            return "B"
        elif average >= 50:
            return "C"
        else:
            return "F"

    def display(self):
        return f"Name: {self.name} Roll: {self.roll}, Average: {self.get_average()}"
```

```python
students = [
    Student("ANkit", 1, [100,45,600]),
    Student("Prajwal", 2, [10,55,60]),
    Student("Anuj", 3, [100,50,60]),
    Student("Jenish", 4, [10,50,63]),
    Student("Prabesh", 5, [20,58,60]),
]
print(students[0].get_average())
students[0].is_pass()
```

```python
predicted_grades = []

for student in students:
    predicted_grades.append(student.predict_grade())

print(predicted_grades)
```

```python
results = [s.predict_grade() for s in students]

print(results)
```
