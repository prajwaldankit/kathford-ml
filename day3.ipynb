```python
def check_even(num):
    try:
        num = int(num)
    except:
        print("INvalid data")
        return
    
    if num % 2 == 0:
        print("even")
    else:
        print("odd")

num = input("Enter a number")
check_even(num)
```

```python
def fibonacci(n):
    result = []
    last , current = 1,2
    result.append(last)
    result.append(current)
    for _ in range(1,n-1):
        new_item = last + current
        last, current = current, new_item
        result.append(new_item)



    return result
fibonacci(5)
```

```python
def rec_fib(n):
    if n <= 0:
        return []
    if n == 1:
        return [1]
    if n == 2:
        return [1, 2]
    series = rec_fib(n - 1)
    series.append(series[-1] + series[-2])
    return series

rec_fib(12)
```
