```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)
y1 = np.cos(x)
y2 = np.tan(x)

fig, ax = plt.subplots()
ax.set_ylim((-2,2))
ax.plot(x, y)
ax.plot(x, y1)
ax.plot(x, y2)
plt.show()
```

```python
x1 = np.linspace(0,50,100)
y1 = np.random.rand(100)

fig,ax = plt.subplots()

ax.scatter(x1,y1)
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.plot(x,y)
plt.title("example plot")
plt.xlabel("time")
plt.ylabel("score")
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,5,6,7,3])

plt.scatter(x,y)
plt.show()
```

```python
years = ["2021", "2022", "2023", "2024"]
gdp = [1000,1010,900,1400]

plt.bar(years,gdp)
plt.show()
```

```python
x = np.linspace(0,50,60)
random_noise = np.random.rand(60)
y = 3 * x + random_noise * 10
y2 = 2.9* x + 7 
plt.scatter(x,y)
plt.plot(x,y2)
plt.xlim((0,50))
plt.ylim((0,200))
plt.show()
```
